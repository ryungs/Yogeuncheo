```python
import pandas as pd
```

```python
origin_data = pd.read_csv('./last_data.txt', sep=",")
origin_data
```

```python
elementary_school_data = pd.read_csv('./학교/초등학교.txt')
elementary_school_data['gu'] = elementary_school_data.apply(lambda x:x["address"].split()[1], axis=1)
elementary_school_data['dong'] = elementary_school_data.apply(lambda x:x["address"].split()[2], axis=1)

middle_school_data = pd.read_csv('./학교/중학교.txt')
middle_school_data['gu'] = middle_school_data.apply(lambda x:x["address"].split()[1], axis=1)
middle_school_data['dong'] = middle_school_data.apply(lambda x:x["address"].split()[2], axis=1)

high_school_data = pd.read_csv('./학교/고등학교.txt')
high_school_data['gu'] = high_school_data.apply(lambda x:x["address"].split()[1], axis=1)
high_school_data['dong'] = high_school_data.apply(lambda x:x["address"].split()[2], axis=1)
```

```python
new_elementary = elementary_school_data.groupby("dong")
new_elementary = pd.DataFrame(new_elementary["address"].count())
new_elementary.columns = ["elementary_count"]

new_middle = middle_school_data.groupby("dong")
new_middle = pd.DataFrame(new_middle["address"].count())
new_middle.columns = ["middle_count"]

new_high = high_school_data.groupby("dong")
new_high = pd.DataFrame(new_high["address"].count())
new_high.columns = ["high_count"]
```

```python
origin_data = pd.merge(origin_data,new_elementary,on="dong" ,how="left").fillna(0)
origin_data["elementary_count"] = origin_data["elementary_count"].apply(lambda x:int(x))

origin_data = pd.merge(origin_data,new_middle,on="dong" ,how="left").fillna(0)
origin_data["middle_count"] = origin_data["middle_count"].apply(lambda x:int(x))

origin_data = pd.merge(origin_data,new_high,on="dong" ,how="left").fillna(0)
origin_data["high_count"] = origin_data["high_count"].apply(lambda x:int(x))
origin_data
```

```python
park_data = pd.read_csv('./공원/공원.txt')
park_data['gu'] = park_data.apply(lambda x:x["address"].split()[1], axis=1)
park_data['dong'] = park_data.apply(lambda x:x["address"].split()[2], axis=1)
park_data
```

```python
new_park = park_data.groupby("dong")
new_park = pd.DataFrame(new_park["address"].count())
new_park.columns = ["park_count"]
```

```python
origin_data = pd.merge(origin_data,new_park,on="dong" ,how="left").fillna(0)
origin_data["park_count"] = origin_data["park_count"].apply(lambda x:int(x))
origin_data
```

```python
station_data = pd.read_csv('./역/지하철.txt')
station_data['gu'] = station_data.apply(lambda x:x["address"].split()[1], axis=1)
station_data['dong'] = station_data.apply(lambda x:x["address"].split()[2], axis=1)
station_data
```

```python
korea_data = pd.read_csv('../data_gdr/1-2. 서울 법정동 주소목록(최종).txt', sep="\t")
korea_data["gu"] = korea_data.apply(lambda x:x["법정동명"].split()[1] if len(x["법정동명"].split())==3 else "",axis=1)
korea_data["dong"] = korea_data.apply(lambda x:x["법정동명"].split()[2] if len(x["법정동명"].split())==3 else "",axis=1)
korea_data.rename(columns={"법정동명":"address"}, inplace = True)
korea_data
```

```python
arr = []
for i in korea_data['dong']:
    input_arr = []
    for j in range(len(station_data)):
        if station_data['dong'][j] == i:
            input_arr.append(station_data['title'][j])
    if len(input_arr)>0:
        arr.append(input_arr)
    else:
        arr.append([])
```

```python
arr
```

```python
origin_data['station'] = arr
origin_data
```

```python
for idx in range(len(origin_data)):
    if origin_data['station'][idx]:
        origin_data['station'][idx] = str(origin_data['station'][idx])
        origin_data['station'][idx] = origin_data['station'][idx][1:-1]
    else:
        origin_data['station'][idx] = '없음'
```
