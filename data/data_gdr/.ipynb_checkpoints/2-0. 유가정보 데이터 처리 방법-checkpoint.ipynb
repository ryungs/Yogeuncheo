### 2. 유가 정보를 pandas로 불러옵니다.

2-1. 유가 정보(10월 4주차).txt를 tap기준으로 자른후 oli data에 저장합니다.

이제 유가 정보는 oil data 입니다.

```python
import pandas as pd
oil_data = pd.read_csv('./2-3. 유가 정보(10월 4주차) 주소변환.txt', sep="\t")
oil_data
```

### 3. 유가 정보 중 동을 기준으로 고급휘발유, 휘발유, 경유 평균 가격 구하기

이제 oil_data 중 동을 기준으로 고급휘발유, 휘발유, 경유의 평균을 구합니다. 

평균을 구하기 위해 먼저 가격이 0인 데이터를 NaN으로 변환합니다.

이때 numpy를 이용합니다.

```python
import numpy as np
oil_data = oil_data.replace(0, np.NaN)
oil_data
```

동을 기준으로 고급휘발유, 휘발유, 경유의 평균을 담은 데이터는 

oil_avg_price 입니다

```python
oil_avg_price = oil_data.groupby(['동'], as_index=False).mean()
oil_avg_price
```

### 4. 평균 가격이 담긴 데이터 txt 파일로 저장하기

```python
oil_avg_price = pd.DataFrame(oil_avg_price)
oil_avg_price.to_csv('./2-4. 유가 정보(10월 4주차) 평균 가격(최종).txt', header=True, index=False, encoding="UTF-8")
```
